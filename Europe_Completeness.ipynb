```python
# Loading necessary data packages
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
import numpy as np
from tqdm.notebook import tqdm

# Importing necessary libraries
# sys.path.append('') #'your-path-to/osm-flex/src'

import osm_flex
import osm_flex.download as dl
import osm_flex.extract as ex
import osm_flex.config
import osm_flex.clip as cp

osm_flex.enable_logs()
```

1. Using the OSM Flex package to download the Road Network for Specified Countries

```python
# List of ISO3 country codes
countries = ['GBR','ITA', 'ESP', 'UKR', 'POL', 'ROU', 'NLD', 'BEL', 'SWE', 'CZE']

roadnetworks = {}

for iso3 in tqdm (countries, desc='Running'):
    print(f"Downloading data for {iso3}...")
    path = dl.get_country_geofabrik(iso3)
    print(f'Saved {iso3} data to: {path}')

    # Extract main roads and keep only LineStrings
    gdf_mainroad = ex.extract_cis(path, 'main_road')
    gdf_mainroad = gdf_mainroad[gdf_mainroad.geometry.geom_type == 'LineString']

    print(f'{iso3}: extracted {len(gdf_mainroad)} main roads as LineStrings.\n')

    roadnetworks[iso3] = gdf_mainroad
    print(f'Saved {iso3} data to: {path}')

    # Extract main roads and keep only LineStrings
    gdf_mainroad = ex.extract_cis(path, 'main_road')
    gdf_mainroad = gdf_mainroad[gdf_mainroad.geometry.geom_type == 'LineString']

    print(f'{iso3}: extracted {len(gdf_mainroad)} main roads as LineStrings.\n')

    roadnetworks[iso3] = gdf_mainroad
```

2. Plot Completeness per Country

```python
country_names = {
    'GBR': 'United Kingdom',
    'ITA': 'Italy',
    'ESP': 'Spain',
    'UKR': 'Ukraine',
    'POL': 'Poland',
    'ROU': 'Romania',
    'NLD': 'Netherlands',
    'BEL': 'Belgium',
    'SWE': 'Sweden',
    'CZE': 'Czech Republic'
}

# Analyze attribute completeness and plot results
columns = ['maxspeed', 'lanes', 'surface']
summary_data = {}

for iso3, gdf_roads in roadnetworks.items():
    if gdf_roads.empty:
        print(f"No road data for {iso3}, skipping.")
        continue

    # Select columns that exist
    available_cols = [col for col in columns if col in gdf_roads.columns]
    if not available_cols:
        print(f"No relevant attributes found in {iso3}, skipping.")
        continue

    completeness = gdf_roads[available_cols].notnull().mean() * 100
    summary_data[iso3] = completeness

# Plot completeness per country
    plt.figure(figsize=(8, 5))
    completeness.plot(kind='bar', color='skyblue', edgecolor='black')
    # Get the full country name using the iso3 code from the dictionary
    full_country_name = country_names.get(iso3, iso3)
    plt.title(f'{full_country_name} Road Attribute Completeness (%)')
    plt.ylabel('Percentage')
    plt.ylim(0, 100)
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    for idx, val in enumerate(completeness):
        plt.text(idx, val + 1, f'{val:.1f}%', ha='center', va='bottom')

    plt.tight_layout()
    plt.show()
```

3. Summarize Results

```python
# Create summary DataFrame and display
summary_df = pd.DataFrame(summary_data).T.fillna(0)
print("Summary completeness table (in %):")
display(summary_df.round(1))
```

4. Plot combined Graph

```python
country_names = {
    'ESP': 'Spain',
    'GBR': 'United Kingdom',
    'ITA': 'Italy',
    'UKR': 'Ukraine',
    'POL': 'Poland',
    'ROU': 'Romania',
    'NLD': 'Netherlands',
    'BEL': 'Belgium',
    'SWE': 'Sweden',
    'CZE': 'Czech Republic'
}

# Plot combined grouped bar chart
df_completeness = summary_df[columns]

# Reorder the DataFrame to put Spain first and the rest alphabetically
other_countries = [country for country in df_completeness.index if country != 'ESP']
# Sort the remaining countries alphabetically based on their full names
other_countries_sorted = sorted(other_countries, key=lambda iso3: country_names.get(iso3, iso3))
ordered_countries = ['ESP'] + other_countries_sorted
df_completeness = df_completeness.reindex(ordered_countries)


fig, ax = plt.subplots(figsize=(14, 7))
bar_width = 0.25
indices = np.arange(len(df_completeness))

for i, col in enumerate(columns):
    ax.bar(indices + i * bar_width, df_completeness[col], width=bar_width,
           label=col, edgecolor='black')

# Get the full country names based on the ISO3 codes in the index
full_country_labels = [country_names.get(iso3, iso3) for iso3 in df_completeness.index]

ax.set_xticks(indices + bar_width)
ax.set_xticklabels(full_country_labels, rotation=45, ha='right')
ax.set_ylim(0, 110)
ax.set_ylabel('Completeness (%)')
ax.set_title('Road Attribute Completeness by European Country')
ax.legend(title='Attributes')
ax.grid(axis='y', linestyle='--', alpha=0.7)

for i in range(len(df_completeness)):
    for j, col in enumerate(columns):
        val = df_completeness.iloc[i, j]
        ax.text(i + j * bar_width, val + 2, f'{val:.1f}%', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```
